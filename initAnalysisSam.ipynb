```python


# Assume df is your DataFrame
from itertools import islice

def plot_race_path(df, game_col="SESSION_GUID", 
                   x_col="M_WORLDPOSITIONX_1", 
                   y_col="M_WORLDPOSITIONY_1", 
                   color_col="M_SPEED_1"):
    """
    Plots the path taken in each game session, colored by a chosen stat.
    """
    # Group by game
    for game, gdf in islice(df.groupby(game_col), 5, 11):
        plt.figure(figsize=(10, 8))
        sc = plt.scatter(
            gdf[x_col], gdf[y_col],
            c=gdf[color_col], cmap="viridis", s=10, alpha=0.8
        )
        plt.colorbar(sc, label=color_col)
        plt.title(f"Racing Path - {game}")
        plt.xlabel("World X Position")
        plt.ylabel("World Z Position")
        plt.axis("equal")
        plt.show()

# Example usage:
plot_race_path(data)
```

```python
import matplotlib.pyplot as plt
import pandas as pd

def plot_game_paths(df, game_col="SESSION_GUID", 
                    x_col="M_WORLDPOSITIONX_1", 
                    y_col="M_WORLDPOSITIONY_1", 
                    color_col="M_SECTOR_1"):
    """
    Plots race paths per game/session, coloured by a chosen variable (default: sector).
    """
    for game, gdf in df.groupby(game_col):
        plt.figure(figsize=(10, 8))
        
        # Draw continuous line so path is outlined
        sc = plt.scatter(
            gdf[x_col], gdf[y_col],
            c=gdf[color_col], cmap="plasma",
            s=6, alpha=0.8
        )
        plt.plot(gdf[x_col], gdf[y_col], color="lightgrey", linewidth=0.5, alpha=0.6)
        
        plt.colorbar(sc, label=color_col)
        plt.title(f"Game path: {game}")
        plt.xlabel("World X Position")
        plt.ylabel("World Z Position")
        plt.axis("equal")
        plt.show()
```

```python
plot_game_paths(data)
```
